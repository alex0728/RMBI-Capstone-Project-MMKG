```python
import pandas as pd
FB15K = pd.read_csv('D:\RMBI_MMKG_data\FB15K\FB15K_EntityTriples.csv')
FB15K.drop('ImageID', inplace=True, axis=1)
FB15K
```

```python
FB15K = FB15K.drop_duplicates(subset=['LabelName'])
FB15K
```

```python
import json
import urllib

img_des = pd.DataFrame(columns=['LabelName','name','description'])

for index, row in FB15K.iterrows():
    api_key = "AIzaSyB4rJ3gScCzeY2Offyc0fsI9UpwsddWlAw"
    mid = row['LabelName']
    service_url = 'https://kgsearch.googleapis.com/v1/entities:search'
    params = {
        'ids': mid,
        'limit': 10,
        'indent': True,
        'key': api_key,
    }
    url = service_url + '?' + urllib.parse.urlencode(params)
    response = json.loads(urllib.request.urlopen(url).read())
    for element in response['itemListElement']:
        try:
            r={'LabelName': mid,'name': element['result']['name'],'description': element['result']['detailedDescription']['articleBody']}
            img_des = img_des.append(r, ignore_index = True)
        except:
            r={'LabelName': mid,'name': element['result']['name'],'description': ''}
            img_des = img_des.append(r, ignore_index = True)
```
